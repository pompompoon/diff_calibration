```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import umap
import seaborn as sns
import matplotlib as mpl
import japanize_matplotlib

# ファイルパスの設定
path = "G:/共有ドライブ/GAP_長寿研/user/iwamoto/視線の動きの俊敏さ/data"
file_path = f"{path}/cyozyuken_b.csv"

# 結果を保存するディレクトリ
import os
result_path = r"G:\共有ドライブ\GAP_長寿研\user\iwamoto\視線の動きの俊敏さ\result"
os.makedirs(result_path, exist_ok=True)

# データの読み込み
# データはCSVなのでsep=','を指定 (必要に応じてsep='\t'に変更)
data = pd.read_csv(file_path, sep=',')

# データの基本情報を確認
print("データのサイズ:", data.shape)
print("カラム名:", data.columns[:10], "...")  # 最初の10カラムのみ表示

# EyeSideとMoCAを除いた数値データのみを特徴量として使用
# 特徴量行列の作成
X = data.drop(['EyeSide', 'MoCA','InspectionDateAndId','Target'], axis=1, errors='ignore')

# NaN値の確認と処理（必要に応じて）
print("NaN値の合計:", X.isna().sum().sum())
if X.isna().sum().sum() > 0:
    X = X.fillna(X.mean())  # NaN値を平均値で置換

# 特徴量のスケーリング（標準化）
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# UMAPによる次元削減
# n_neighbors: 近傍点の数（小さいと局所構造を保持、大きいとグローバル構造を保持）
# min_dist: 埋め込み後の点の最小距離（小さいとクラスターが密になる）
# n_components: 次元削減後の次元数
reducer = umap.UMAP(n_neighbors=15, min_dist=0.1, n_components=2, random_state=42)
embedding = reducer.fit_transform(X_scaled)

# K-meansクラスタリングを適用
# シルエットスコアを使用して最適なクラスタ数を決定
silhouette_scores = []
max_clusters = min(10, len(embedding) - 1)  # データ数より少ないクラスタ数に制限
k_range = range(2, max_clusters + 1)

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    cluster_labels = kmeans.fit_predict(embedding)
    score = silhouette_score(embedding, cluster_labels)
    silhouette_scores.append(score)
    print(f"クラスタ数 {k}: シルエットスコア {score:.4f}")

# 最適なクラスタ数
optimal_k = k_range[np.argmax(silhouette_scores)]
print(f"最適なクラスタ数: {optimal_k}")
optimal_k = 4
# 最適なクラスタ数でK-meansを実行
kmeans = KMeans(n_clusters=optimal_k, random_state=42, n_init=10)
cluster_labels = kmeans.fit_predict(embedding)

# クラスタの中心点
cluster_centers = kmeans.cluster_centers_

# 結果をデータフレームに格納
result_df = pd.DataFrame({
    'UMAP1': embedding[:, 0],
    'UMAP2': embedding[:, 1],
    'EyeSide': data['EyeSide'] if 'EyeSide' in data.columns else None,
    'MoCA': data['MoCA'] if 'MoCA' in data.columns else None,
    'Cluster': cluster_labels
})

print("UMAP結果:")
print(result_df)

# シルエットスコアの可視化
plt.figure(figsize=(10, 6))
plt.plot(k_range, silhouette_scores, 'o-')
plt.xlabel('クラスタ数')
plt.ylabel('シルエットスコア')
plt.title('クラスタ数によるシルエットスコアの変化')
plt.grid(True)
plt.tight_layout()
plt.savefig(os.path.join(result_path, 'silhouette_scores.png'))
plt.show()

# MoCAによる色分け可視化
plt.figure(figsize=(10, 8))
sns.scatterplot(
    x='UMAP1', 
    y='UMAP2', 
    hue='MoCA' if 'MoCA' in data.columns else None,
    palette='viridis',
    data=result_df
)
plt.title('UMAP次元削減結果（MoCAによる色分け）')
plt.tight_layout()
plt.savefig(os.path.join(result_path, 'umap_result_moca.png'))
plt.show()

# クラスタによる色分け可視化
plt.figure(figsize=(10, 8))
scatter = sns.scatterplot(
    x='UMAP1', 
    y='UMAP2', 
    hue='Cluster',
    palette='tab10',
    data=result_df
)
# クラスタの中心点をプロット
plt.scatter(
    cluster_centers[:, 0],
    cluster_centers[:, 1],
    s=100,
    c='black',
    marker='X',
    label='Cluster Centers'
)
plt.title(f'UMAP次元削減結果（K-means: {optimal_k}クラスタ）')
plt.legend(title='Cluster')
plt.tight_layout()
plt.savefig(os.path.join(result_path, 'umap_result_kmeans.png'))
plt.show()

# 3次元に削減した場合の可視化とK-means（オプション）
if False:  # 3次元可視化が必要な場合はFalseをTrueに変更
    reducer_3d = umap.UMAP(n_neighbors=15, min_dist=0.1, n_components=3, random_state=42)
    embedding_3d = reducer_3d.fit_transform(X_scaled)
    
    # 3次元空間でK-meansクラスタリング
    kmeans_3d = KMeans(n_clusters=optimal_k, random_state=42, n_init=10)
    cluster_labels_3d = kmeans_3d.fit_predict(embedding_3d)
    
    result_df_3d = pd.DataFrame({
        'UMAP1': embedding_3d[:, 0],
        'UMAP2': embedding_3d[:, 1],
        'UMAP3': embedding_3d[:, 2],
        'EyeSide': data['EyeSide'] if 'EyeSide' in data.columns else None,
        'MoCA': data['MoCA'] if 'MoCA' in data.columns else None,
        'Cluster': cluster_labels_3d
    })
    
    from mpl_toolkits.mplot3d import Axes3D
    
    # MoCAによる色分け
    fig = plt.figure(figsize=(12, 10))
    ax = fig.add_subplot(111, projection='3d')
    
    scatter = ax.scatter(
        embedding_3d[:, 0],
        embedding_3d[:, 1],
        embedding_3d[:, 2],
        c=data['MoCA'] if 'MoCA' in data.columns else None,
        cmap='viridis',
        s=50
    )
    
    if 'MoCA' in data.columns:
        plt.colorbar(scatter, ax=ax, label='MoCA')
        
    ax.set_title('UMAP 3D Visualization (MoCA)')
    ax.set_xlabel('UMAP1')
    ax.set_ylabel('UMAP2')
    ax.set_zlabel('UMAP3')
    
    plt.tight_layout()
    plt.savefig(os.path.join(result_path, 'umap_3d_result_moca.png'))
    plt.show()
    
    # クラスタリング結果による色分け
    fig = plt.figure(figsize=(12, 10))
    ax = fig.add_subplot(111, projection='3d')
    
    scatter = ax.scatter(
        embedding_3d[:, 0],
        embedding_3d[:, 1],
        embedding_3d[:, 2],
        c=cluster_labels_3d,
        cmap='tab10',
        s=50
    )
    
    plt.colorbar(scatter, ax=ax, label='Cluster')
        
    ax.set_title(f'UMAP 3D Visualization (K-means: {optimal_k} clusters)')
    ax.set_xlabel('UMAP1')
    ax.set_ylabel('UMAP2')
    ax.set_zlabel('UMAP3')
    
    plt.tight_layout()
    plt.savefig(os.path.join(result_path, 'umap_3d_result_kmeans.png'))
    plt.show()

# クラスタごとの特徴分析
cluster_stats = result_df.groupby('Cluster').describe()
print("\nクラスタ統計情報:")
print(cluster_stats)

# クラスタごとの元データの特徴量平均値を計算
original_features = X.columns
cluster_means = pd.DataFrame()
for cluster in range(optimal_k):
    cluster_indices = result_df[result_df['Cluster'] == cluster].index
    cluster_mean = X.iloc[cluster_indices].mean()
    cluster_means[f'Cluster_{cluster}'] = cluster_mean

# 各クラスタの特徴の比較をヒートマップで可視化
plt.figure(figsize=(16, 12))
sns.heatmap(cluster_means, annot=False, cmap='coolwarm', center=0)
plt.title('各クラスタにおける特徴量の平均値ヒートマップ')
plt.tight_layout()
plt.savefig(os.path.join(result_path, 'cluster_features_heatmap.png'))
plt.show()

# 結果をCSVファイルとして保存
result_csv_path = os.path.join(result_path, 'umap_result.csv')
result_df.to_csv(result_csv_path, index=False)
print(f"UMAP結果を{result_csv_path}に保存しました")

# クラスタの特徴量平均値をCSVに保存
cluster_means_csv_path = os.path.join(result_path, 'cluster_means.csv')
cluster_means.to_csv(cluster_means_csv_path)
print(f"クラスタごとの特徴量平均値を{cluster_means_csv_path}に保存しました")
```
