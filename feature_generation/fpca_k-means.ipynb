```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D
import matplotlib as mpl
import japanize_matplotlib

# ファイルパスの設定
path = "G:/共有ドライブ/GAP_長寿研/user/iwamoto/視線の動きの俊敏さ/data"
file_path = f"{path}/shizuoka_0327_a-1st.csv"

# 結果を保存するディレクトリ
result_path = r"G:\共有ドライブ\GAP_長寿研\user\iwamoto\視線の動きの俊敏さ\result"
os.makedirs(result_path, exist_ok=True)

# データの読み込み
# データがカンマ区切りの場合はsep=','、タブ区切りの場合はsep='\t'
data = pd.read_csv(file_path, sep=',')

# データの基本情報を確認
print("データのサイズ:", data.shape)
print("カラム名の一部:", data.columns[:10], "...")  # 最初の10カラムのみ表示

# EyeSideとMoCAを除いた数値データのみを特徴量として使用
# 特徴量行列の作成
X = data.drop(['EyeSide', 'MoCA','InspectionDateAndId','Target'], axis=1, errors='ignore')

# NaN値の確認と処理
nan_count = X.isna().sum().sum()
print(f"\nNaN値の合計: {nan_count}")
if nan_count > 0:
    print("NaN値を含むカラム:")
    print(X.columns[X.isna().any()].tolist())
    X = X.fillna(X.mean())  # NaN値を平均値で置換

# 特徴量のスケーリング（標準化）
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# ----- 1. PCAによる次元削減 -----
# 説明変数の95%を保持する主成分数を自動的に決定
pca = PCA(n_components=0.95, random_state=42)
X_pca = pca.fit_transform(X_scaled)

print(f"元の特徴量数: {X_scaled.shape[1]}")
print(f"PCA後の特徴量数: {X_pca.shape[1]}")
print(f"累積寄与率: {np.sum(pca.explained_variance_ratio_):.4f}")

# PCAの各主成分の寄与率をプロット
plt.figure(figsize=(10, 6))
plt.bar(range(1, len(pca.explained_variance_ratio_) + 1), pca.explained_variance_ratio_)
plt.plot(range(1, len(pca.explained_variance_ratio_) + 1), 
         np.cumsum(pca.explained_variance_ratio_), 'r-o')
plt.xlabel('主成分')
plt.ylabel('寄与率・累積寄与率')
plt.title('PCA: 各主成分の寄与率と累積寄与率')
plt.grid(True)
plt.tight_layout()
plt.savefig(os.path.join(result_path, 'pca_variance_ratio.png'))

# ----- 2. K-meansクラスタリング -----
# シルエットスコアを使用して最適なクラスタ数を決定
silhouette_scores = []
max_clusters = min(10, len(X_pca) - 1)  # データ数より少ないクラスタ数に制限
k_range = range(2, max_clusters + 1)

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    cluster_labels = kmeans.fit_predict(X_pca)
    score = silhouette_score(X_pca, cluster_labels)
    silhouette_scores.append(score)
    print(f"クラスタ数 {k}: シルエットスコア {score:.4f}")

# 最適なクラスタ数
optimal_k = k_range[np.argmax(silhouette_scores)]
print(f"最適なクラスタ数: {optimal_k}")
optimal_k = 4
# 最適なクラスタ数でK-meansを実行
kmeans = KMeans(n_clusters=optimal_k, random_state=42, n_init=10)
cluster_labels = kmeans.fit_predict(X_pca)

# クラスタの中心点（主成分空間）
cluster_centers_pca = kmeans.cluster_centers_

# ----- 可視化 -----

# シルエットスコアの可視化
plt.figure(figsize=(10, 6))
plt.plot(k_range, silhouette_scores, 'o-')
plt.xlabel('クラスタ数')
plt.ylabel('シルエットスコア')
plt.title('クラスタ数によるシルエットスコアの変化')
plt.grid(True)
plt.tight_layout()
plt.savefig(os.path.join(result_path, 'silhouette_scores.png'))

# PCAの2次元可視化（最初の2つの主成分）
plt.figure(figsize=(12, 10))

# 2次元散布図にクラスタを色分けして表示
plt.scatter(
    X_pca[:, 0], 
    X_pca[:, 1], 
    c=cluster_labels, 
    cmap='viridis', 
    alpha=0.7, 
    s=50
)

# クラスタの中心点をプロット
plt.scatter(
    cluster_centers_pca[:, 0], 
    cluster_centers_pca[:, 1], 
    c='red', 
    marker='X', 
    s=200, 
    label='Cluster Centers'
)

plt.xlabel('第1主成分 (PC1)')
plt.ylabel('第2主成分 (PC2)')
plt.title(f'PCA + K-means: {optimal_k}クラスタ')
plt.colorbar(label='クラスタ')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig(os.path.join(result_path, 'pca_kmeans_2d.png'))

# 3次元可視化（最初の3つの主成分がある場合）
if X_pca.shape[1] >= 3:
    fig = plt.figure(figsize=(14, 12))
    ax = fig.add_subplot(111, projection='3d')
    
    # 3次元散布図にクラスタを色分けして表示
    scatter = ax.scatter(
        X_pca[:, 0],
        X_pca[:, 1],
        X_pca[:, 2],
        c=cluster_labels,
        cmap='viridis',
        alpha=0.7,
        s=50
    )
    
    # クラスタの中心点をプロット
    ax.scatter(
        cluster_centers_pca[:, 0],
        cluster_centers_pca[:, 1],
        cluster_centers_pca[:, 2],
        c='red',
        marker='X',
        s=200,
        label='Cluster Centers'
    )
    
    ax.set_xlabel('第1主成分 (PC1)')
    ax.set_ylabel('第2主成分 (PC2)')
    ax.set_zlabel('第3主成分 (PC3)')
    ax.set_title(f'PCA + K-means: {optimal_k}クラスタ (3D)')
    plt.colorbar(scatter, ax=ax, label='クラスタ')
    ax.legend()
    plt.tight_layout()
    plt.savefig(os.path.join(result_path, 'pca_kmeans_3d.png'))

# ----- 結果の集約 -----

# MoCAの値がある場合、クラスタごとのMoCA統計を計算
if 'MoCA' in data.columns:
    moca_cluster_df = pd.DataFrame({
        'MoCA': data['MoCA'],
        'Cluster': cluster_labels
    })
    
    # クラスタごとのMoCA平均値
    moca_cluster_stats = moca_cluster_df.groupby('Cluster')['MoCA'].describe()
    print("\nクラスタごとのMoCA統計:")
    print(moca_cluster_stats)
    
    # クラスタごとのMoCA分布をボックスプロットで可視化
    plt.figure(figsize=(12, 6))
    sns.boxplot(x='Cluster', y='MoCA', data=moca_cluster_df)
    plt.title('クラスタごとのMoCAスコア分布')
    plt.grid(True)
    plt.tight_layout()
    plt.savefig(os.path.join(result_path, 'moca_by_cluster.png'))

# EyeSideの値がある場合、クラスタとEyeSideのクロス集計
if 'EyeSide' in data.columns:
    eyeside_cluster_df = pd.DataFrame({
        'EyeSide': data['EyeSide'],
        'Cluster': cluster_labels
    })
    
    # クラスタとEyeSideのクロス集計
    eyeside_cluster_cross = pd.crosstab(
        eyeside_cluster_df['Cluster'], 
        eyeside_cluster_df['EyeSide'], 
        normalize='index'
    ) * 100  # パーセンテージに変換
    
    print("\nクラスタとEyeSideのクロス集計（%）:")
    print(eyeside_cluster_cross)
    
    # クラスタとEyeSideの関係を積み上げ棒グラフで可視化
    plt.figure(figsize=(10, 6))
    eyeside_cluster_cross.plot(kind='bar', stacked=True)
    plt.title('クラスタごとのEyeSide割合')
    plt.xlabel('クラスタ')
    plt.ylabel('割合 (%)')
    plt.grid(True)
    plt.tight_layout()
    plt.savefig(os.path.join(result_path, 'eyeside_by_cluster.png'))

# クラスタごとの元の特徴量の平均値を計算
cluster_feature_means = pd.DataFrame()
for cluster in range(optimal_k):
    cluster_indices = np.where(cluster_labels == cluster)[0]
    cluster_mean = pd.DataFrame(X.iloc[cluster_indices].mean()).T
    cluster_mean.index = [f'Cluster_{cluster}']
    cluster_feature_means = pd.concat([cluster_feature_means, cluster_mean])

# クラスタごとの元特徴量の平均値をヒートマップで可視化
plt.figure(figsize=(20, max(8, optimal_k * 0.8)))
sns.heatmap(
    cluster_feature_means,
    annot=False,
    cmap='coolwarm',
    center=0,
    yticklabels=cluster_feature_means.index
)
plt.title('各クラスタにおける特徴量の平均値')
plt.tight_layout()
plt.savefig(os.path.join(result_path, 'cluster_feature_means_heatmap.png'))

# 主成分の負荷量（元の特徴量との相関）を計算
loadings = pca.components_
feature_names = X.columns

# 主成分の負荷量をデータフレームに変換
n_pcs_to_show = min(5, len(loadings))
loadings_df = pd.DataFrame(
    data=loadings[:n_pcs_to_show].T,
    index=feature_names,
    columns=[f'PC{i+1}' for i in range(n_pcs_to_show)]
)
"""
plt.figure(figsize=(10, max(10, len(feature_names) * 0.4)))  # サイズを大きく
sns.heatmap(loadings_df, 
            annot=True,  # 値を表示
            fmt=".2f",   # 小数点以下2桁で表示
            cmap='coolwarm', 
            center=0,
            annot_kws={"size": 8},  # アノテーションのフォントサイズ
            linewidths=0.5)  # セル間の線
plt.title('主成分の負荷量（元の特徴量との相関）', fontsize=14)  # タイトルのフォントサイズ
plt.xticks(fontsize=10)  # X軸ラベルのフォントサイズ
plt.yticks(fontsize=10)  # Y軸ラベルのフォントサイズ
plt.tight_layout()
plt.savefig(os.path.join(result_path, 'pca_loadings_heatmap.png'), dpi=300, bbox_inches='tight')  # 解像度を上げる
# ----- 結果の保存 -----
"""
# 負荷量のヒートマップ - 横長表示のための修正
plt.figure(figsize=(22, 10))  # 横幅を大きく、縦幅を適度に
sns.heatmap(
    loadings_df.T,  # 転置して主成分を行、特徴量を列に
    annot=False,    # 値が多すぎるので注釈は表示しない
    cmap='coolwarm',
    center=0,
    yticklabels=True,  # Y軸ラベル（主成分）を表示
    xticklabels=True   # X軸ラベル（特徴量）を表示
)
plt.title('上位主成分と元の特徴量の関係')
plt.xticks(rotation=90)  # X軸ラベル（特徴量名）を縦に回転
plt.tight_layout()
plt.savefig(os.path.join(result_path, 'pca_loadings_heatmap.png'), dpi=300, bbox_inches='tight')

# 主成分とクラスタラベルを含むデータフレームを作成
result_df = pd.DataFrame(
    data=X_pca,
    columns=[f'PC{i+1}' for i in range(X_pca.shape[1])]
)
result_df['Cluster'] = cluster_labels

# 元のラベル情報があれば追加
if 'EyeSide' in data.columns:
    result_df['EyeSide'] = data['EyeSide'].values
if 'MoCA' in data.columns:
    result_df['MoCA'] = data['MoCA'].values

# 結果をCSVに保存
result_csv_path = os.path.join(result_path, 'pca_kmeans_results.csv')
result_df.to_csv(result_csv_path, index=False)
print(f"\nPCA+K-means結果を{result_csv_path}に保存しました")

# 主成分の負荷量をCSVに保存
loadings_csv_path = os.path.join(result_path, 'pca_loadings.csv')
loadings_df.to_csv(loadings_csv_path)
print(f"PCA負荷量を{loadings_csv_path}に保存しました")

# クラスタごとの特徴量平均値をCSVに保存
cluster_means_csv_path = os.path.join(result_path, 'cluster_feature_means.csv')
cluster_feature_means.to_csv(cluster_means_csv_path)
print(f"クラスタごとの特徴量平均値を{cluster_means_csv_path}に保存しました")

print("\nPCA+K-means分析が完了しました。")
```
