```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os

# Specified paths
output_dir = r"G:\共有ドライブ\GAP_長寿研\user\iwamoto\視線の動きの俊敏さ"
csv_file_path = r"G:\共有ドライブ\GAP_長寿研\user\iwamoto\視線の動きの俊敏さ\data\calibration_20853_20240201_20240719094849.csv"

# Read the CSV predictions file
def load_predictions(file_path):
    try:
        df = pd.read_csv(file_path)
        print(f"Successfully loaded CSV from {file_path}")
        print(f"Shape: {df.shape}")
        print("Columns:", df.columns.tolist())
        return df
    except Exception as e:
        print(f"Error loading CSV file: {e}")
        return None

# Visualize eye tracking from CSV data
def plot_eye_tracking(df, x_col, y_col, time_col, save_path=None):
    # Filter out invalid data points if any
    filtered_df = df.dropna(subset=[x_col, y_col, time_col])
    
    # Set up the figure
    plt.figure(figsize=(12, 8))
    
    # Create a scatter plot with points colored by timestamp
    scatter = plt.scatter(
        filtered_df[x_col], 
        filtered_df[y_col],
        c=filtered_df[time_col], 
        cmap='viridis',
        alpha=0.7,
        s=30
    )
    
    # Add a colorbar to show the time progression
    cbar = plt.colorbar(scatter)
    cbar.set_label('Time')
    
    # Add a line connecting the points in sequence
    plt.plot(
        filtered_df[x_col], 
        filtered_df[y_col],
        'r-', 
        alpha=0.3,
        linewidth=1
    )
    
    # Set labels and title
    plt.xlabel(f'{x_col} (degrees)')
    plt.ylabel(f'{y_col} (degrees)')
    plt.title('Eye Movement Trajectory')
    
    # Add grid for better readability
    plt.grid(True, alpha=0.3)
    
    # Make equal aspect ratio
    plt.axis('equal')
    
    # Invert Y-axis if appropriate for eye tracking
    plt.gca().invert_yaxis()
    
    plt.tight_layout()
    
    # Save the figure if a path is provided
    if save_path:
        plt.savefig(save_path, dpi=300, bbox_inches='tight')
        print(f"Figure saved to {save_path}")
    
    plt.show()

# Function to create a heatmap of eye fixations
def plot_eye_heatmap(df, x_col, y_col, resolution=100, save_path=None):
    # Filter out invalid data points
    filtered_df = df.dropna(subset=[x_col, y_col])
    
    # Create a 2D histogram
    x = filtered_df[x_col]
    y = filtered_df[y_col]
    
    # Calculate appropriate ranges for the heatmap
    x_min, x_max = x.min() - 1, x.max() + 1
    y_min, y_max = y.min() - 1, y.max() + 1
    
    # Create the heatmap
    plt.figure(figsize=(12, 8))
    plt.hist2d(x, y, bins=resolution, cmap='hot', range=[[x_min, x_max], [y_min, y_max]])
    plt.colorbar(label='Count')
    
    plt.xlabel(f'{x_col} (degrees)')
    plt.ylabel(f'{y_col} (degrees)')
    plt.title('Eye Fixation Heatmap')
    
    # Make equal aspect ratio
    plt.axis('equal')
    
    # Invert Y-axis if appropriate for eye tracking
    plt.gca().invert_yaxis()
    
    plt.tight_layout()
    
    # Save the figure if a path is provided
    if save_path:
        plt.savefig(save_path, dpi=300, bbox_inches='tight')
        print(f"Heatmap saved to {save_path}")
    
    plt.show()

# Analyze the predictions and plot relevant metrics
def analyze_predictions(df, save_dir=None):
    # Display basic statistics
    print("\nBasic statistics of the data:")
    print(df.describe())
    
    # Plot histograms of numeric columns
    numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns
    
    # Plot histograms for each numeric column
    for col in numeric_cols:
        plt.figure(figsize=(10, 6))
        plt.hist(df[col].dropna(), bins=30, alpha=0.7, color='steelblue')
        plt.title(f'Distribution of {col}')
        plt.xlabel(col)
        plt.ylabel('Frequency')
        plt.grid(True, alpha=0.3)
        
        if save_dir:
            save_path = os.path.join(save_dir, f'{col}_histogram.png')
            plt.savefig(save_path, dpi=300, bbox_inches='tight')
            print(f"Histogram saved to {save_path}")
        
        plt.show()
    
    # If the dataframe has prediction-related columns, create scatter plots
    if 'prediction' in df.columns or any('prediction' in col.lower() for col in df.columns):
        # Find prediction column
        pred_col = [col for col in df.columns if 'prediction' in col.lower()][0]
        
        # Find features or target columns to plot against predictions
        other_cols = [col for col in numeric_cols if col != pred_col][:5]  # Limit to 5 columns
        
        for col in other_cols:
            plt.figure(figsize=(10, 6))
            plt.scatter(df[col], df[pred_col], alpha=0.7)
            plt.title(f'{col} vs {pred_col}')
            plt.xlabel(col)
            plt.ylabel(pred_col)
            plt.grid(True, alpha=0.3)
            
            if save_dir:
                save_path = os.path.join(save_dir, f'{col}_vs_{pred_col}_scatter.png')
                plt.savefig(save_path, dpi=300, bbox_inches='tight')
                print(f"Scatter plot saved to {save_path}")
            
            plt.show()

# Main function
def main():
    # Create output directory if it doesn't exist
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)
        print(f"Created output directory: {output_dir}")
    
    try:
        # Load the CSV data
        df = load_predictions(csv_file_path)
        
        if df is not None:
            print("\nFirst few rows of the data:")
            print(df.head())
            
            # Check if specific eye tracking columns exist in the data
            eye_angle_x_col = None
            eye_angle_y_col = None
            time_col = None
            
            # Try to find appropriate columns for eye movement tracking
            possible_x_cols = ['EyeCenterAngleX', 'eye_x', 'x_angle', 'x']
            possible_y_cols = ['EyeCenterAngleY', 'eye_y', 'y_angle', 'y']
            possible_time_cols = ['TimeStamp', 'time', 'timestamp']
            
            for col in df.columns:
                col_lower = col.lower()
                if any(x_col.lower() in col_lower for x_col in possible_x_cols):
                    eye_angle_x_col = col
                elif any(y_col.lower() in col_lower for y_col in possible_y_cols):
                    eye_angle_y_col = col
                elif any(time_col.lower() in col_lower for time_col in possible_time_cols):
                    time_col = col
            
            if eye_angle_x_col and eye_angle_y_col and time_col:
                print(f"\nFound eye tracking columns: {eye_angle_x_col}, {eye_angle_y_col}, {time_col}")
                
                # Plot eye tracking trajectory
                trajectory_plot_path = os.path.join(output_dir, 'eye_trajectory_plot.png')
                plot_eye_tracking(df, eye_angle_x_col, eye_angle_y_col, time_col, save_path=trajectory_plot_path)
                
                # Plot eye fixation heatmap
                heatmap_plot_path = os.path.join(output_dir, 'eye_fixation_heatmap.png')
                plot_eye_heatmap(df, eye_angle_x_col, eye_angle_y_col, save_path=heatmap_plot_path)
            else:
                print("\nEye tracking columns not found. Available columns:")
                print(df.columns.tolist())
                print("\nPerforming general analysis of the CSV data instead.")
            
            # Analyze the predictions
            analyze_predictions(df, save_dir=output_dir)
            
    except Exception as e:
        print(f"An error occurred: {e}")
        import traceback
        traceback.print_exc()

if __name__ == "__main__":
    main()
```
