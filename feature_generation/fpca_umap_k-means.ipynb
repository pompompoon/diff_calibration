```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import umap
import seaborn as sns
import matplotlib as mpl
import japanize_matplotlib

# ファイルパスの設定
path = "G:/共有ドライブ/GAP_長寿研/user/iwamoto/視線の動きの俊敏さ/data"
file_path = f"{path}/cyozyuken_b.csv"

# 結果を保存するディレクトリ
result_path = r"G:\共有ドライブ\GAP_長寿研\user\iwamoto\視線の動きの俊敏さ\result"
os.makedirs(result_path, exist_ok=True)

# データの読み込み
# データがカンマ区切りの場合はsep=','、タブ区切りの場合はsep='\t'
data = pd.read_csv(file_path, sep=',')

# データの基本情報を確認
print("データのサイズ:", data.shape)
print("カラム名:", data.columns[:10], "...")  # 最初の10カラムのみ表示

# EyeSideとMoCAを除いた数値データのみを特徴量として使用
# 特徴量行列の作成
X = data.drop(['EyeSide', 'MoCA','InspectionDateAndId','Target'], axis=1, errors='ignore')

# NaN値の確認と処理（必要に応じて）
print("NaN値の合計:", X.isna().sum().sum())
if X.isna().sum().sum() > 0:
    X = X.fillna(X.mean())  # NaN値を平均値で置換

# 特徴量のスケーリング（標準化）
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# ----- 1. PCAによる次元削減 -----
# 説明変数の95%を保持する主成分数を自動的に決定
pca = PCA(n_components=0.95, random_state=42)
X_pca = pca.fit_transform(X_scaled)

print(f"元の特徴量数: {X_scaled.shape[1]}")
print(f"PCA後の特徴量数: {X_pca.shape[1]}")
print(f"累積寄与率: {np.sum(pca.explained_variance_ratio_):.4f}")

# PCAの各主成分の寄与率をプロット
plt.figure(figsize=(10, 6))
plt.bar(range(1, len(pca.explained_variance_ratio_) + 1), pca.explained_variance_ratio_)
plt.plot(range(1, len(pca.explained_variance_ratio_) + 1), 
         np.cumsum(pca.explained_variance_ratio_), 'r-o')
plt.xlabel('主成分')
plt.ylabel('寄与率・累積寄与率')
plt.title('PCA: 各主成分の寄与率と累積寄与率')
plt.grid(True)
plt.tight_layout()
plt.savefig(os.path.join(result_path, 'pca_variance_ratio.png'))

# 最初の2つの主成分で可視化
plt.figure(figsize=(10, 8))
plt.scatter(X_pca[:, 0], X_pca[:, 1], alpha=0.7)
plt.xlabel('第1主成分')
plt.ylabel('第2主成分')
plt.title('PCA: 最初の2つの主成分での可視化')
plt.grid(True)
plt.tight_layout()
plt.savefig(os.path.join(result_path, 'pca_scatter.png'))

# ----- 2. UMAPによる次元削減 -----
# PCA後のデータにUMAPを適用
reducer = umap.UMAP(n_neighbors=15, min_dist=0.1, n_components=2, random_state=42)
embedding = reducer.fit_transform(X_pca)

# ----- 3. K-meansクラスタリング -----
# シルエットスコアを使用して最適なクラスタ数を決定
silhouette_scores = []
max_clusters = min(10, len(embedding) - 1)  # データ数より少ないクラスタ数に制限
k_range = range(2, max_clusters + 1)

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    cluster_labels = kmeans.fit_predict(embedding)
    score = silhouette_score(embedding, cluster_labels)
    silhouette_scores.append(score)
    print(f"クラスタ数 {k}: シルエットスコア {score:.4f}")

# 最適なクラスタ数
optimal_k = k_range[np.argmax(silhouette_scores)]
print(f"最適なクラスタ数: {optimal_k}")
optimal_k = 4
# 最適なクラスタ数でK-meansを実行
kmeans = KMeans(n_clusters=optimal_k, random_state=42, n_init=10)
cluster_labels = kmeans.fit_predict(embedding)

# クラスタの中心点
cluster_centers = kmeans.cluster_centers_

# 結果をデータフレームに格納
result_df = pd.DataFrame({
    'UMAP1': embedding[:, 0],
    'UMAP2': embedding[:, 1],
    'EyeSide': data['EyeSide'] if 'EyeSide' in data.columns else None,
    'MoCA': data['MoCA'] if 'MoCA' in data.columns else None,
    'Cluster': cluster_labels
})

print("UMAP結果:")
print(result_df.head())

# ----- 可視化 -----
# シルエットスコアの可視化
plt.figure(figsize=(10, 6))
plt.plot(k_range, silhouette_scores, 'o-')
plt.xlabel('クラスタ数')
plt.ylabel('シルエットスコア')
plt.title('クラスタ数によるシルエットスコアの変化')
plt.grid(True)
plt.tight_layout()
plt.savefig(os.path.join(result_path, 'silhouette_scores.png'))

# MoCAによる色分け可視化
plt.figure(figsize=(10, 8))
if 'MoCA' in data.columns:
    sns.scatterplot(
        x='UMAP1', 
        y='UMAP2', 
        hue='MoCA',
        palette='viridis',
        data=result_df
    )
    plt.title('PCA + UMAP次元削減結果（MoCAによる色分け）')
else:
    sns.scatterplot(
        x='UMAP1', 
        y='UMAP2',
        data=result_df
    )
    plt.title('PCA + UMAP次元削減結果')
plt.tight_layout()
plt.savefig(os.path.join(result_path, 'pca_umap_result_moca.png'))

# クラスタによる色分け可視化
plt.figure(figsize=(10, 8))
scatter = sns.scatterplot(
    x='UMAP1', 
    y='UMAP2', 
    hue='Cluster',
    palette='tab10',
    data=result_df
)
# クラスタの中心点をプロット
plt.scatter(
    cluster_centers[:, 0],
    cluster_centers[:, 1],
    s=100,
    c='black',
    marker='X',
    label='Cluster Centers'
)
plt.title(f'PCA + UMAP次元削減結果（K-means: {optimal_k}クラスタ）')
plt.legend(title='Cluster')
plt.tight_layout()
plt.savefig(os.path.join(result_path, 'pca_umap_result_kmeans.png'))

# ----- クラスタ分析 -----
# クラスタごとの特徴分析
if 'MoCA' in data.columns:
    cluster_stats = result_df.groupby('Cluster')['MoCA'].describe()
    print("\nクラスタごとのMoCA統計情報:")
    print(cluster_stats)

# クラスタごとの元データの特徴量平均値を計算
original_features = X.columns
cluster_means = pd.DataFrame()
for cluster in range(optimal_k):
    cluster_indices = result_df[result_df['Cluster'] == cluster].index
    cluster_mean = X.iloc[cluster_indices].mean()
    cluster_means[f'Cluster_{cluster}'] = cluster_mean

# 各クラスタの特徴の比較をヒートマップで可視化
plt.figure(figsize=(16, 12))
sns.heatmap(cluster_means, annot=False, cmap='coolwarm', center=0)
plt.title('各クラスタにおける特徴量の平均値ヒートマップ')
plt.tight_layout()
plt.savefig(os.path.join(result_path, 'cluster_features_heatmap.png'))

# PCAの主成分とオリジナル特徴量の関係性を調べる
pca_components = pd.DataFrame(
    pca.components_,
    columns=X.columns,
    index=[f'PC{i+1}' for i in range(pca.n_components_)]
)

# 上位5つの主成分について、特徴量との相関をヒートマップで可視化
plt.figure(figsize=(16, 8))
top_n_components = min(5, pca.n_components_)
sns.heatmap(
    pca_components.iloc[:top_n_components],
    cmap='coolwarm',
    center=0,
    annot=False
)
plt.title('上位主成分と元の特徴量の関係')
plt.tight_layout()
plt.savefig(os.path.join(result_path, 'pca_components_heatmap.png'))

# ----- 結果の保存 -----
# UMAP結果をCSVファイルとして保存
result_csv_path = os.path.join(result_path, 'pca_umap_result.csv')
result_df.to_csv(result_csv_path, index=False)
print(f"PCA+UMAP結果を{result_csv_path}に保存しました")

# クラスタの特徴量平均値をCSVに保存
cluster_means_csv_path = os.path.join(result_path, 'cluster_means.csv')
cluster_means.to_csv(cluster_means_csv_path)
print(f"クラスタごとの特徴量平均値を{cluster_means_csv_path}に保存しました")

# PCA主成分の係数をCSVに保存
pca_components_csv_path = os.path.join(result_path, 'pca_components.csv')
pca_components.to_csv(pca_components_csv_path)
print(f"PCA主成分の係数を{pca_components_csv_path}に保存しました")
```
